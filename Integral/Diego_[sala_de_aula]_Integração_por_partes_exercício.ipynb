```python
# Questão 1
# Enunciado:
# Use integração por partes para calcular a integral indefinida
#     ∫ x e^{2x} dx
# Depois, faça o gráfico da função f(x) = x e^{2x} em um intervalo adequado e discuta o comportamento da função.


import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = x*sp.exp(2*x)

F = sp.integrate(f, x)
print("Integral indefinida ∫ x e^{2x} dx =")
sp.pprint(F)

x_vals = np.linspace(-1, 1, 400)
f_lamb = sp.lambdify(x, f, 'numpy')
y_vals = f_lamb(x_vals)

plt.figure(figsize=(7,4))
plt.plot(x_vals, y_vals, label=r'$f(x)=x e^{2x}$')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title(r'Função $f(x)=x e^{2x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Questão 2
# Enunciado:
# Calcule a integral indefinida
#     ∫ x^2 cos(x) dx
# usando integração por partes duas vezes.
# Em seguida, represente graficamente a função f(x) = x^2 cos(x) em um intervalo que inclua algumas oscilações.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f1 = x**2 * sp.cos(x)
F1 = sp.integrate(f1, x)

print("Integral indefinida ∫ x^2 cos(x) dx =")
sp.pprint(F1)
f_lamb = sp.lambdify(x, f1, 'numpy')

x_vals = np.linspace(-1, 1, 400)
y_vals = f_lamb(x_vals)
plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label=r'$f(x)=x^2 cos(x)$')
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.title(r'Gráfico de $f(x)=x^2 cos(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Questão 3
# Enunciado:
# Use integração por partes para calcular a integral definida
#     ∫_0^2 x ln(x+1) dx
# Mostre também a antiderivada encontrada.
# Depois, faça o gráfico de f(x) = x ln(x+1) e destaque a área sob a curva no intervalo [0, 2].


import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f2 =  x**2 * sp.log(x + 1)
F2 = sp.integrate(f2, x)

print("Integral indefinida ∫_0^2 x ln(x+1) dx =")
sp.pprint(F2)
f_lamb = sp.lambdify(x, f2, 'numpy')

x_vals = np.linspace(0, 2, 400)
y_vals = f_lamb(x_vals)
plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label=r'$f(x)=x^2 cos(x)$')
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.title(r'Gráfico de $f(x)=x^2 cos(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Questão 4
# Enunciado:
# Calcule a integral definida
#     ∫_0^π x sin(x) dx
# utilizando integração por partes.
# Depois, represente graficamente f(x) = x sin(x) e destaque a área sob a curva entre x = 0 e x = π.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

x = sp.symbols('x')
f = x*sp.sin(x)

F = sp.integrate(f, x)
print("Primitiva F(x) de x sin(x):")
sp.pprint(F)

I = sp.integrate(f, (x, 0, sp.pi))
print("\nIntegral definida ∫_0^π x sin(x) dx = ", sp.simplify(I))

f_lamb = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(0, float(sp.pi), 400)
y_vals = f_lamb(x_vals)

fig, ax = plt.subplots(figsize=(7,4))
ax.plot(x_vals, y_vals, 'r', label=r'$f(x)=x\sin x$')
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)

x_fill = np.linspace(0, float(sp.pi), 300)
y_fill = f_lamb(x_fill)
verts = [(0,0), *zip(x_fill, y_fill), (float(sp.pi),0)]
poly = Polygon(verts, facecolor='1.0', edgecolor='0.5', alpha=0.7)
ax.add_patch(poly)

ax.set_title(r'Área de $\int_0^\pi x\sin(x)\,dx$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Questão 5
# Enunciado:
# Lembre que ∫ ln(x) dx pode ser resolvida por integração por partes.
# (a) Encontre uma primitiva de ln(x) usando integração por partes.
# (b) Use essa primitiva para calcular a integral definida
#         ∫_1^3 ln(x) dx
# (c) Faça o gráfico de f(x) = ln(x) e destaque a área sob a curva no intervalo [1, 3].

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

x = sp.symbols('x')
f3 = sp.log(x)

F3 = sp.integrate(f3, x)
print("Primitiva F3(x) de ln(x):")
sp.pprint(f3)

I = sp.integrate(f3, (x, 1, 3))
print("\nIntegral definida ∫_1^3 ln(x) dx = ", sp.simplify(I))

f_lamb = sp.lambdify(x, f3, 'numpy')
x_vals = np.linspace(0.1, 3, 400)  # começa em 0.1 pra evitar ln(0)
y_vals = f_lamb(x_vals)

fig, ax = plt.subplots(figsize=(7,4))
ax.plot(x_vals, y_vals, 'r', label=r'$f(x)=\ln(x)$')

ax.axhline(0, linewidth=0.8)
ax.axvline(0, linewidth=0.8)

x_fill = np.linspace(1, 3, 400)
y_fill = f_lamb(x_fill)
verts = [(1,0), *zip(x_fill, y_fill), (3,0)]
poly = Polygon(verts, facecolor='1.0', edgecolor='0.5', alpha=0.7)
ax.add_patch(poly)

ax.set_title(r'Área de $\int_1^3 \ln(x)\,dx$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```
